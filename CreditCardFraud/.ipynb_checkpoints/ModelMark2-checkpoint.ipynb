```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

Data Pre-processing:

```python
data = pd.read_csv("creditcard.csv")
print(data.head())
print(data.iloc[4308],"\n")
print(data.shape)
```

```python
print(data.describe())
```

Class header represents the state of transaction, (0 = valid, 1 = fraudulent). Since mean is very low (0.001727) this means that there are a low amount of fraudulent transactions

```python
data.hist(figsize = (15,15))
plt.show()
```

As shown in the above histogram for the class header, there are an almost negligble amount of fraudulent transactions in the context of the total amount of transactions.

```python
fraud = data[data["Class"]==1]
valid = data[data["Class"]==0]
```

Lets create a correlation matrix to highlight correlated features in a seaborn heatmap plot

```python
correlation_matrix = data.corr()
fig = plt.figure(figsize = (10,10))
sns.heatmap(correlation_matrix,square = True)
plt.show()
```

```python
columns = data.columns.tolist()
target = "Class"
columns = [c for c in columns if c not in [target]]

X = data[columns]
Y = data[target]

print(X.shape,Y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = .5)
print(Y_train.shape,Y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier

classifier = KNeighborsClassifier()
classifier.fit(X_train,Y_train)
predictions = classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,predictions))
```

As mentioned previously, the amount of fraudulent transactions are almost negligible and therefore can statistically can be classified as outliers in the scope of the entire dataset. Perhaps using an outlier focused algorithm such as an Isolation Forest classifier would be even more effective.

```python
from sklearn.neighbors import LocalOutlierFactor 

#threshold of decision function
thresh = fraud/valid
clf = LocalOutlierFactor(contamination = thresh)
clf.fit_predict(X_train)
```
