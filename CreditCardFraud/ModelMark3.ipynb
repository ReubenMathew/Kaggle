```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("creditcard.csv")
```

```python
print(data.head())
print(data.describe())
print(data.shape)
```

Header Column Class represents the state of transaction (1 = fraudulent, 0 = valid)

```python
sns.countplot(data['Class'])
plt.show()
print(data['Class'].sum())
```

The average for the Class Column head is 0.001727 meaning that a very low amount of transactions in this dataset are considered fraudulent. This could mean using an outlier focused classifier or model would be theoretically effective.

```python
data = data.sample(frac = 0.50, random_state = 1)
```

```python
print(data.shape)
print(data['Class'].sum())
```

A correlation matrix might show us some features to focus on and perhaps increase accuracy and improve speed of classification

```python
correlation_matrix = data.corr()
fig = plt.figure(figsize = (10,10))
sns.heatmap(correlation_matrix,square = True)
plt.show()
```

```python
data = data.drop(['V10','V12','V14','V17'],1)
```

```python
fraud = data[data['Class']==1]
valid = data[data['Class']==0]

percent_fraud = len(fraud)/len(valid)
print(percent_fraud)
```

```python
columns = data.columns.tolist()

target = "Class"

columns = [c for c in columns if c not in [target]]
print(columns)

X = data[columns]
Y = data[target]

from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = .5)
print(Y_train.shape,Y_test.shape)

print(X.shape,Y.shape)
```

Preprocessing Complete!

```python
from sklearn.metrics import classification_report, accuracy_score
from sklearn.ensemble import IsolationForest
```

Lets try an Isolation Forest Classifier as this could be effective when dealing with outlier cases

```python
import warnings
warnings.filterwarnings("ignore")
clf = IsolationForest(max_samples=len(X_train))
clf.fit(X_train)
y_pred = clf.predict(X_test)
y_pred[y_pred==1] = 0
y_pred[y_pred==-1] = 1
```

```python
print(accuracy_score(y_pred,Y_test))
print(y_pred.sum(),Y_test.sum(),(Y_test.sum()/y_pred.sum()))
```

This is not bad but perhaps tweaking the contamination parameter, which represents the proportion of outliers in the dataset, will improve the accuracy of the model. Assigning it to the percentage of fraudulent transactions would be appropriate in this case as this represents the proportion of anomalies.

```python
clf = IsolationForest(contamination = percent_fraud, max_samples = len(X_train))
clf.fit(X_train)
y_pred = clf.predict(X_test)
y_pred[y_pred==1] = 0
y_pred[y_pred==-1] = 1
```

```python
print(accuracy_score(y_pred,Y_test))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_pred,Y_test))
```

```python
print(y_pred.sum(),Y_test.sum(),(Y_test.sum()/y_pred.sum()))
```
